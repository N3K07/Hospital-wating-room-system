```python
class Patient:
  global id
  global name
  global age
  global bloodgroup
  def __init__(self,id,name,age,bloodgroup):
    self.id = id
    self.name = name
    self.age = age
    self.bloodgroup = bloodgroup

class Node:
  def __init__(self,patient):
    self.patient = patient
    self.next = None
    self.prev = None

class WRM:
  def __init__(self):
    self.head = None
    self.tail = None

  def is_empty(self):
    return self.head is None

  def register_patient(self,id,name,age,bloodgroup):
    new_patient = Patient(id,name,age,bloodgroup)
    new_node = Node(new_patient)
    if self.is_empty():
      self.head = new_node
      self.tail = new_node
      new_node.next = new_node
      new_node.prev = new_node
    else:
      new_node.prev = self.tail
      new_node.next = self.head
      self.head.prev = new_node
      self.tail.next = new_node
      self.tail = new_node

  def serve_patient(self):
    if self.is_empty():
      print("There is no patients in the waiting room currently")
    else:
      served_patient = self.head.patient
      if self.head == self.tail:
        self.head = None
        self.tail = None
      else:
        self.head = self.head.next
        self.tail.next = self.head
        self.head.perv = self.tail
      print(f'Currently serving patient: {served_patient.name}')

  def cancel_all(self):
    if self.is_empty():
      print("There is no patients in the waiting room currently")
    else:
      self.head = None
      self.tail = None
      print("Canceled all appointments")

  def can_doctor_go_home(self):
    if self.is_empty():
      print("Yes, doctor can go home")
    else:
      print("No, doctor can't go home right now")

  def show_all_patients(self):
    if self.is_empty():
      print("The appointment list is empty")
    else:
      current_node = self.head
      while True:
        print(f'Paitent ID: {current_node.patient.id}\nPatient Name: {name}\nPatient Age: {age}\nPatient Bloodgroup: {bloodgroup}')
        if current_node == self.tail:
          break
        current_node = current_node.next

wrm = WRM()
while True:
  print("Waiting room management system")
  print('------------------------------')
  print("1. Add patient.\n2. Serve patient.\n3. Show all patients in current list.\n4. Can doctor go home.\n5. Cancel all appointments.\n6. Exit menu.")
  print('------------------------------')
  print("\n")
  inp = input("Enter your service number: \n")

  if inp =='1':
    id = int(input('Enter ID: '))
    name = input('Enter Name: ')
    age = int(input('Enter Age: '))
    bloodgroup = input('Enter Bloodgroup: ')
    wrm.register_patient(id,name,age,bloodgroup)
    print("Added successfully")
  elif inp == '2':
    wrm.serve_patient()
  elif inp == '3':
    wrm.show_all_patients()
  elif inp == '4':
    wrm.can_doctor_go_home()
  elif inp == '5':
    wrm.cancel_all()
  elif inp == '6':
    break
  else:
    print("invalid input")
```
